```python
x = data['Рік']
y = data['Iwt']
plt.hist(x)
plt.show()
```

```python
from scipy import stats
res = stats.linregress(x, y)
plt.plot(x, y, 'o', label='Iwt')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```
