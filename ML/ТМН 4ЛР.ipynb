```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.metrics import confusion_matrix
from yellowbrick.cluster import InterclusterDistance
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train[:5000].reshape(5000, 28*28) # flatten the 28x28 / cut data
y_train = y_train[:5000]
```

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(x_train)
cm = pd.DataFrame(confusion_matrix(y_train, kmeans.labels_))
cm
```

```python
x_train_dn = x_train[:20]
```

```python
hier = linkage(x_train_dn, method = 'ward')
dn = dendrogram(hier)
plt.show()
labels = [y_train[i] for i in dn['leaves']]
dn = dendrogram(hier, labels = labels)
plt.show()
```
