```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line plot

```python
#matplotlib
x = df['carat']
y = df['price']
plt.plot(x, color = 'purple', linewidth=3,)
plt.show()
```

```python
#pandas
df.plot(x='carat', y='price', linewidth=3)
#plt.figure(figsize = (12,4))
plt.show()
```

```python
# Pandas:
df.plot.line(x='cut', y='price', title='Price vs cut', figsize=(10,6))
```

```python
sns.lineplot(x='clarity', y='price', data=df, color="b")
ax2 = plt.twinx()
sns.set(rc={"figure.figsize":(6, 10)})
```

```python
sns.lineplot(x='depth', y='price', data=df, color="orange")
ax2 = plt.twinx()
sns.set(rc={"figure.figsize":(6, 10)})
```

```python
# Matplotlib
ax = plt.gca()
df.plot(kind='line', x='price', y='carat', ax=ax, figsize=(10,6))
df.plot(kind='line', x='price', y='depth', ax=ax, figsize=(10,6))
```

### Histogram

```python
# Pandas
df['cut'].value_counts().sort_index().plot.bar(figsize=(10,6))
```

```python
#Seaborn
sns.countplot(df['cut'])
```

```python
#matplotlib
fig, ax = plt.subplots()
new_data = df['cut'].value_counts().sort_index()
ind = new_data.index
freq = new_data.values
ax.bar(ind, freq)
```

```python
# Pandas
df['clarity'].value_counts().sort_index().plot.bar(figsize=(10,6), color='pink')
```

### Boxplot

```python
# Seaborn
sns.boxplot(y='price', data=df)
```

```python
# Matplotlib
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot()
ax.boxplot(df['price'])
ax.set_ylabel('price')
```

### Violin Plot

```python
#seaborn
sns.violinplot(data=df, x="cut", y="depth", hue="cut")
```

```python
sns.violinplot(data=df, x="color", y="price", inner="stick")
```
